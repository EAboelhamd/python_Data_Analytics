```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('CLV.csv')
df.head()
```

```python
df.corr()
```

```python
df.describe().transpose()
```

```python
train, test = train_test_split(df, test_size=0.2)
```

```python
x = train[['Total Claim Amount', 'Monthly Premium Auto', 'Income', 'Months Since Last Claim', 'Months Since Policy Inception', 'Number of Policies']]
y = train['Customer Lifetime Value']
```

```python
np.random.seed(3)
n =  np.shape(train)[0]
# x = np.random.uniform(-15, 15, size = n)
# y = x**2 + 2*np.random.randn(n, )
X = np.reshape(x ,[n, 6]) 
y = np.reshape(y ,[n ,])
```

```python
clf = MLPRegressor(alpha=0.0001, hidden_layer_sizes = (6,), max_iter = 50000, 
                 activation = 'logistic', verbose = 'True', learning_rate = 'adaptive')
a = clf.fit(X, y)
```

```python
#x_ = np.linspace(-4, 4, 160) # define axis

pred_x = np.reshape(test[['Total Claim Amount', 'Monthly Premium Auto', 'Income', 'Months Since Last Claim', 'Months Since Policy Inception', 'Number of Policies']], [len(test), 6]) # [160, ] -> [160, 1]
pred_y = clf.predict(pred_x) # predict network output given x_
e = np.square(np.subtract(test['Customer Lifetime Value'],pred_y)).mean()
e
# fig = plt.figure() 
# plt.plot(e, color = 'b') # plot original function
# plt.plot(pred_y, test['Customer Lifetime Value'], '-') # plot network output
```
